## Так как у циана изменилась разметка и структура сайта, в проекте будут использоваться данные, распарсенные ранее.

```python
df=pd.read_json('full_data.json')
```

```python
#редактирование колонки price
df['price'] = df['price'].replace(to_replace=r'\s+', value='', regex=True)
df['price'] = df['price'].replace(to_replace=r'\D', value='', regex=True)
df['price'] = pd.to_numeric(df['price'])
```

```python
df.head(3)
```

```python
df['area'] = df['area'].replace(to_replace=r'р-н\s*', value='', regex=True)
```

```python
df.head(3)
```

```python
# разделение столбца title
df[['title', 'square']] = df['title'].str.split(',', 1, expand=True)
```

```python
df.head(3)
```

```python
#обработка столбца square
df['square'] = df['square'].replace(to_replace=r'\s*м²', value='', regex=True)
df['square'] = df['square'].replace(to_replace=r',\d', value='', regex=True)
df['square'] = df['square'].replace(to_replace=r'\s', value='', regex=True)
df['square'] = pd.to_numeric(df['square'])
```

```python
df.head()
```

```python
df['facilities'] = df['facilities'] + df['facilities2']
df = df.drop(['facilities2'], axis=1)
```

```python
df.head()
```

```python
df['title'] = df['title'].replace(to_replace=r'-комн.', value='', regex=True)
df['title'] = df['title'].replace(to_replace=r'квартира', value='', regex=True)
df['title'] = df['title'].replace(to_replace=r'[Сс]тудия',
                                  value='0',
                                  regex=True)
df['title'] = df['title'].replace(to_replace=r'[Аа]партамент[ы-]',
                                  value='',
                                  regex=True)
df['title'] = df['title'].str.strip()
df['title'] = df['title'].replace(to_replace=r'-',
                                  value='',
                                  regex=True)
```

```python
df.title.unique()
```

```python
df.head()
```

```python
df.info()
```

```python
#разделение столбца time_metro и редактирование
df['time_metro'] = df['time_metro'].replace(to_replace=r'мин.',
                                            value='',
                                            regex=True)
df['time_metro'] = df['time_metro'].str.strip()
df['time_metro'] = df['time_metro'].replace(to_replace='  ',
                                            value=' ',
                                            regex=True)
df[['time_metro', 'transport_type']] = df['time_metro'].str.split(' ',
                                                                  1,
                                                                  expand=True)
df['time_metro'] = pd.to_numeric(df['time_metro'])
```

```python
df.head(4)
```

```python
df.transport_type.unique()
```

```python
df['floor_sqare'] = [','.join(map(str, l)) for l in df['floor_sqare']]
```

```python
df['floor_sqare'].value_counts()
```

```python
df['floor_sqare'] = [','.join(map(str, l)) for l in df['floor_sqare']]
df['floor_sqare'] = df['floor_sqare'].str.extract(r'(\d{1,2}\sиз\s\d{1,2})')
df[['floor', 'final_floor']] = df['floor_sqare'].str.split('из', 1, expand=True)
df = df.drop(['floor_sqare'], axis=1)
```

```python
df = df[(df.title != 'Многокомнатная') | (df.title != 'Многокомнатные')]
```

```python
df.title.unique()
```

```python
df.info()
```

```python
df.head()
```

```python
df.transport_type.unique()
```

```python
df['floor'] = pd.to_numeric(df['floor'])
df['final_floor'] = pd.to_numeric(df['final_floor'])
```

```python
df['title'] = df['title'].replace(['Многокомнатная', 'Многокомнатные'], 6)
```

```python
df['time_metro'] = df['time_metro'].astype('int64')
```

```python
df['time_metro'] = df['time_metro'].fillna(df['time_metro'].mean())
```

```python
df.to_json('cian_data.json', orient='records')
```
